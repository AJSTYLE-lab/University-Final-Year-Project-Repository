filling Nan values with average of integer for int/float column data type and mode with string or object data type:

```python
for i in file.columns:
  if file[i].dtype=="int64" or file[i].dtype=="float64":
    file[i] = file[i].fillna(file[i].mean())
  elif file[i].dtype=="object":
    file[i] = file[i].fillna(file[i].mode()[0])
file.drop_duplicates(inplace=True)
file.shape
```

```python
new_file=pd.read_csv("/content/DENGUE_FILE_FINAL.csv")
new_file.head()
```

```python
new_file.shape
```

```python
new_file.dtypes
```

```python
new_file.isnull().sum()
```

```python

import numpy as np

new_file['Value'] = new_file['Value'].replace('-', np.nan)
new_file['NumValue'] = new_file['NumValue'].replace('-', np.nan)
new_file['Value'] = pd.to_numeric(new_file['Value'], errors='coerce')
new_file['NumValue'] = pd.to_numeric(new_file['NumValue'], errors='coerce')
new_file.dtypes
```

```python
for i in new_file.columns:
  if new_file[i].dtype=="int64" or new_file[i].dtype=="float64":
    new_file[i]=new_file[i].fillna(new_file[i].mean())
  elif new_file[i].dtype=="object":
    new_file[i] =new_file[i].fillna(new_file[i].mode()[0])
new_file.drop_duplicates(inplace=True)
new_file.isnull().sum()
```

```python
new_file.head()
```

```python
# @title Population vs Time

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(new_file['Population'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(new_file, x='Time', y='Population', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title Time vs NumValue

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Time']
  ys = series['NumValue']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = new_file.sort_values('Time', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Population')):
  _plot_series(series, series_name, i)
  fig.legend(title='Population', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Time')
_ = plt.ylabel('NumValue')
```
